### Implementation

Use the code cell (or multiple code cells, if necessary) to implement the first step of your project. Once you have completed your implementation and are satisfied with the results, be sure to thoroughly answer the questions that follow.

````python
import tensorflow as tf
from sklearn.utils import shuffle
from scipy import ndimage
from random import randint

### Features and labels
x = tf.placeholder(tf.float32, (None, 32, 32, 3))
y = tf.placeholder(tf.int32, (None))
one_hot_y = tf.one_hot(y, 43)
````

````python
### Preprocessing data
# Balance out number of examples per label
X_train_extra = X_train
y_train_extra = y_train

for i in count_of_labels:
    # Ensure minimum 2000 examples per label
    if count_of_labels[i] < 2000:
        missing = 2000 - count_of_labels[i]
        
        sample = X_train[np.where(y_train == i)]
        X_new = []
        y_new = []       
        
        for n in range (missing):
            index = randint(0, len(sample)-1)
            angle = randint(-30, 30)
            X_new.append(ndimage.rotate(sample[index], angle, reshape=False))
            y_new.append(i)
            
        X_train_extra = np.append(X_train_extra, X_new, axis=0)
        y_train_extra = np.append(y_train_extra, y_new, axis=0)

# Normalize to 0.1 - 0.9 
X_train_normalize_extra = X_train_extra / 255 * 0.8 + 0.1

# Check ranges
print(X_train_extra.min(), X_train_extra.max(), X_train_normalize_extra.min(), X_train_normalize_extra.max())        

# Split training set to train and validation sets
from sklearn.cross_validation import train_test_split

X_train, X_validation, y_train, y_validation = train_test_split(X_train_normalize_extra, y_train_extra, 
                                                                test_size=0.20, random_state=1000)

del missing, sample, X_new, y_new, index, angle, X_train_extra, X_train_normalize_extra, y_train_extra
````

````python
### Visualize processed data
# Create a dictionary to hold sample size for each label
count_of_labels_extra = {}
y_list = y_train.tolist()

# Loop through labels to count distinct signs and plot one sample for each
for i in np.unique(y_train):
    count_of_labels_extra[i] = y_list.count(i)
    for n in range(len(y_train)):
        if y_train[n] == i:
            plt.subplot(5, 10, i+1)
            plt.imshow(X_train[n], cmap="gray")
            plt.axis('off')
            plt.title(y_train[n])
            break 

del y_list
````

````python
# Plot sample size of each label in the training set
plt.bar(range(len(count_of_labels_extra)), count_of_labels_extra.values(), align='center')
plt.title("Sample size of each label")
plt.show()
print (count_of_labels_extra)

del count_of_labels, count_of_labels_extra
````

#### Testing process

````python
### Hyperparameters
EPOCHS = 25
BATCH_SIZE = 32
mu = 0
sigma = 0.1
strides = 1
k = 2
rate = 0.001
````

````python
### Implement LeNet-5
from tensorflow.contrib.layers import flatten

# Conv2D wrapper, with bias and relu activation
def conv2d(x, W, b, strides):
    x = tf.nn.conv2d(x, W, strides=[1, strides, strides, 1], padding='VALID')
    x = tf.nn.bias_add(x, b)
    return tf.nn.relu(x)

# Max-pooling
def maxpool2d(x, k):
    return tf.nn.max_pool(x, ksize=[1, k, k, 1], strides=[1, k, k, 1], padding='VALID')

def LeNet(x):        
    # Layer 1
    conv1 = conv2d(x, tf.Variable(tf.truncated_normal(shape=(5, 5, 3, 6), mean = mu, stddev = sigma)), 
                   tf.Variable(tf.zeros(6)), strides)
    conv1 = maxpool2d(conv1, k)

    # Layer 2
    conv2 = conv2d(conv1, tf.Variable(tf.truncated_normal(shape=(5, 5, 6, 16), mean = mu, stddev = sigma)),
                   tf.Variable(tf.zeros(16)), strides)
    conv2 = maxpool2d(conv2, k)
    
    # Fully connected layer
    # Reshape conv3 output to fit fully connected layer input
    fc11 = flatten(conv2)
    fc11 = tf.add(tf.matmul(fc11, tf.Variable(tf.truncated_normal(shape=(400, 120), mean = mu, stddev = sigma))),
                            tf.Variable(tf.zeros(120)))
    fc11 = tf.nn.relu(fc11)
    
    # Fully connected layer
    fc12 = tf.add(tf.matmul(fc11, tf.Variable(tf.truncated_normal(shape=(120, 84), mean = mu, stddev = sigma))),
                            tf.Variable(tf.zeros(84)))
    fc12 = tf.nn.relu(fc12)

    # Output Layer - class prediction
    out = tf.add(tf.matmul(fc12, tf.Variable(tf.truncated_normal(shape=(84, 43), mean = mu, stddev = sigma))),
                           tf.Variable(tf.zeros(43)))
    return out
````

````python
### Train the Model - LeNet
# Training Pipeline
logits = LeNet(x)
cross_entropy = tf.nn.softmax_cross_entropy_with_logits(logits, one_hot_y)
loss_operation = tf.reduce_mean(cross_entropy)
optimizer = tf.train.AdamOptimizer(learning_rate = rate)
training_operation = optimizer.minimize(loss_operation)

# Model Evaluation
correct_prediction = tf.equal(tf.argmax(logits, 1), tf.argmax(one_hot_y, 1))
accuracy_operation = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))
saver = tf.train.Saver()

# Model Evaluation
def evaluate(X_data, y_data):
    num_examples = len(X_data)
    total_accuracy = 0
    sess = tf.get_default_session()
    for offset in range(0, num_examples, BATCH_SIZE):
        batch_x, batch_y = X_data[offset:offset+BATCH_SIZE], y_data[offset:offset+BATCH_SIZE]
        accuracy = sess.run(accuracy_operation, feed_dict={x: batch_x, y: batch_y})
        total_accuracy += (accuracy * len(batch_x))
    return total_accuracy / num_examples

with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    num_examples = len(X_train)
    train_accuracy = []
    validation_accuracy = []
    
    for i in range(EPOCHS):
        X_data, y_data = shuffle(X_train, y_train)
#        log_start = 0 
#        train_accuracy_batch = []
#        validation_accuracy_batch = []
        for offset in range(0, num_examples, BATCH_SIZE):
            end = offset + BATCH_SIZE
            batch_x, batch_y = X_data[offset:end], y_data[offset:end]
            sess.run(training_operation, feed_dict={x: batch_x, y: batch_y})
            
#            if log_start < 100: # log 100 batches
#                train_accuracy_batch.append(evaluate(batch_x, batch_y))
                
#                index = randint(0, int(len(y_validation)-1-BATCH_SIZE*0.2))
#                end = index+int(BATCH_SIZE*0.2)
#                validation_accuracy_batch.append(evaluate(X_validation_normalize[index:end], y_validation[index:end]))
                
#                log_start += 1
            
        train_accuracy.append(evaluate(X_train, y_train))
        validation_accuracy.append(evaluate(X_validation, y_validation))
     
#        plt.plot(range(log_start), train_accuracy_batch, 'r--', label='Training Accuracy')
#        plt.plot(range(log_start), validation_accuracy_batch, 'r', label='Validation Accuracy')
#        plt.legend(loc=4)
#        plt.show()   
        
    saver.save(sess, 'Traffic_Sign_Classifier_final')
    print("Model saved") 

    plt.plot(range(EPOCHS), train_accuracy, 'r', label='Training Epochs Accuracy')
    plt.plot(range(EPOCHS), validation_accuracy, 'x', label='Validation Epochs Accuracy')
    plt.legend(loc=4)
    plt.show()  
    print (train_accuracy, validation_accuracy)
````

````python
### Evaluate the Model
# Normalize test set
X_test_normalize = X_test/ 255 * 0.8 + 0.1

with tf.Session() as sess:
    saver.restore(sess, tf.train.latest_checkpoint('.'))

    test_accuracy = evaluate(X_test_normalize, y_test)
    print("Test Accuracy = {:.3f}".format(test_accuracy))
````

- Activation 1: use ReLU activation.
- Pooling layer 1: use max pooling with ksize and stides equal to 2, the output shape is 14x14x6.

- Activation 2: use ReLU activation.

- Flatten layer: flatten the output shape of the final pooling layer such that it's 1D instead of 3D. 

- Fully connected layer 1: it has 120 outputs.
- Activation 3: use ReLU activation.
- Fully connected layer 2: it has 84 outputs.
- Activation 4: use ReLU activation.

- Fully connected layer 3: this is the output layer. It has 43 outputs.

### Question 4

_How did you train your model? (Type of optimizer, batch size, epochs, hyperparameters, etc.)_


**Answer:**

My testing process is guided by this paper: https://arxiv.org/pdf/1412.6980v8.pdf

- Optimizer: Adam optimizer
    * Adam optimizer uses moving averages of the parameters, it generally performs better than the Gradient Decent Optimizer. Test confirms this.
    
- Learning rate: 0.001
    * Paper suggested Adam optimizer works better with smaller learning rate. I tested few values range(0.001 - 0.01).

- Batch size: 32
    * Larger batch size yield faster computation but worse performance (since epochs was fixed). Tested 10, 32(default suggested by the paper), 100, 500, 1000 and choose 32. 
    * By-batch learning process was plotted and observed to help with the testing

- EPOCHS: 25
    * Started with 50, algorithm learns very fast after just a few epochs, so 25 is selected.

Since these parameters have implication on each other, several combinations were tested before choosing the final set. 

- Other parameters: mu = 0, sigma = 0.1, strides = 1, k = 2
    * Did not change given the initial setting, achieved high accuracy.
